# **Data Cleaning and Pre-processing**


**Handling Missing values**


```python
df.isnull().sum()
```

**Filling Missing values**


```python
df["actual"] = df["actual"].fillna(df["actual"].mean())
df["rfs"] = df["rfs"].fillna(df["rfs"].mean())
df["normal"] = df["normal"].fillna(df["normal"].mean())
df["deviation"] = df["deviation"].fillna(df["deviation"].mean())
```

```python
df.isnull().sum()
```

The dataset contains missing values in columns such as actual rainfall, normal rainfall, and deviation. These missing values were handled using mean imputation, where missing entries were replaced with the average value of the respective column. This approach helps in maintaining data consistency without losing significant information.

**Handling Duplicates**

```python

df.duplicated().sum()
```

The dataset was checked for duplicate records using the duplicated() function. The result showed that there are no duplicate entries in the dataset, ensuring data integrity and reliability for further analysis.

**Rename columns**


```python

df.columns
```

```python
df.rename(columns={
    "state_code" : "state_code",
    "actual" : "actual_rainfall",
    "rfs" : "rainfall_score",
    "normal" : "normal_rainfall",
    "deviation" : "rainfall_deviation"
}, inplace=True)
```

```python
df.columns
```

Column names were renamed to improve clarity and readability. For example, 'actual' was renamed to 'actual_rainfall', 'normal' to 'normal_rainfall', and 'deviation' to 'rainfall_deviation'.

```python
df.head()
```

**Dropping Column "id"**


```python
df.drop(columns=["id"], inplace=True)
```

```python
df.head()
```

The 'id' column was removed as it does not contribute to the analysis and only represents a unique identifier.

**Outlier Detection**


```python
import matplotlib.pyplot as plt

df.boxplot(column=["actual_rainfall", "normal_rainfall", "rainfall_deviation"])
plt.show()
```

Outlier analysis was performed using boxplots. A few outliers were observed in the rainfall-related columns; however, they were retained as they represent natural variations in rainfall patterns and are important for analysis.

**Skewness Check**


```python

df[["actual_rainfall", "normal_rainfall", "rainfall_deviation"]].skew()
```

Skewness was analyzed for rainfall-related variables, and it was observed that the data is positively skewed. Since rainfall data naturally contains extreme values, no transformation was applied to maintain the original data distribution.

**Changing Data Types**


```python
df.dtypes
```

```python
df["date"] =  pd.to_datetime(df["date"])
```

```python
df.dtypes
```

The 'date' column was converted from object type to datetime format to enable time-based analysis.

**Feature Transformation**


```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
```

```python
df.head()
```

New features such as 'year' and 'month' were extracted from the date column to support time-based analysis of rainfall trends

# **Bar Plot: Top 10 States by Average Rainfall**

```python
#Calculate average rainfall by state

state_avg=df.groupby("state_name")["actual_rainfall"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
state_avg.plot(kind="bar")
plt.title("Top 10 States by Average Rainfall")
plt.xlabel("State")
plt.ylabel("Average Rainfall")
plt.xticks(rotation=45)
plt.show()
```

### **Features Used**

* state_name
* actual_rainfall

### **Insights**

* Identifies top 10 states with highest average rainfall.
* Clear variation in rainfall levels across states.

### **Business Interpretation:**

* High rainfall states are suitable for water-intensive crops and hydropower.
* Low rainfall regions may need irrigation and water conservation strategies.
* Helps governments in resource allocation and regional planning.
